# Формирование модели монетизации

## Описание проекта <a id="step1"></a>   
</div>

Наша игра уже пользуется большой популярностью — настало время на ней зарабатывать. 

Задача - сформировать модель монетизации сервиса, а именно:
- Провести исследовательский анализ данных;
- Проанализировать влияние событий на совершение целевого события;
- Проверить статистические гипотезы.

    *Проверить гипотезу различия времени прохождения уровня между пользователями, которые заканчивают уровень через реализацию проекта, и пользователями, которые заканчивают уровень победой над другим игроком.*

## Описание данных <a id="step2"></a>   
</div>

Основной датасет содержит данные о событиях, совершенных в мобильной игре "Космические братья". В ней пользователи строят свою космическую программу и пытаются преуспеть в нелёгком деле колонизации галактики.

Основная монетизация игры — только планируется. Но предполагается, что в приложении будет происходить показ рекламы на экране с выбором типа объекта для постройки.

В датасете представлены данные по игре пользователей на первом уровне. Завершение первого уровня требует от игрока выполнения одного из двух условий:

- Победа над первым врагом;
- Реализация проекта - разработка орбитальной сборки спутников.

В датасете содержатся данные первых пользователей приложения — когорты пользователей, которые начали пользоваться приложением в период с 4 по 10 мая включительно.

**Датасет *events.csv* содержит колонки:**

- `event_datetime` - время совершения события;
- `event` - название события;
- `building_type` - тип объекта строительства;
- `user_id` - идентификатор пользователя;
- `project_type` - тип реализованного проекта.

Расшифровки событий:

- Колонка `event`:
    - `building` - Постройка объекта;
    - `finished_stage_1` - Окончание первого уровня;
    - `project` - Реализация проекта.
- Колонка `building_type` - тип объекта для постройки:
    - `assembly_shop` - Сборочный цех;
    - `spaceport` - Космопорт;
    - `research_center` - Исследовательский центр.

**Датасет *ad_cost.csv* содержит колонки:**

- `day` - день, в который был совершен клик по объявлению;
- `source` - источник трафика;
- `cost` - стоимость кликов.

**Датасет user_source.csv содержит колонки:**

- `user_id` - идентификатор пользователя;
- `source` - источников, с которого пришёл пользователь, установивший приложение.

## Декомпозиция задачи<a id="step3"></a>   
</div>

**Цель исследования:** формирование модели монетизации - а именно, составить прогноз получения прибыли с одного игрока (в зависимости от стиля игры), рассчитать LTV и выдать рекомендации по стратегии монетизации сервиса.

1. Загрузка, первичный осмотр и предобработка данных;


2. Проведение исследовательского анализа данных (EDA);
    - Датасет **events.csv**:
        - Есть ли в наборе данных отсутствующие признаки, изучим средние значения и стандартные отклонения;
        - Посмотрим на распределение количества пользователей по колонкам `event` (название события), `building_type` (тип объекта строительства), `project_type` (тип реализованного проекта) - построим круговые диаграммы;
        - посчитаем средние значения затраченного времени на совершение различных типов `event` (название события), `project_type` (тип реализованного проекта);
        - Сформируем выводы.
    
    - Датасет **ad_cost.csv**:
        - Есть ли в наборе данных отсутствующие признаки, изучим средние значения и стандартные отклонения;
        - Посмотрим на распределение количества кликов по дням недели - построим столбчатую диаграмму;
        - посмотрим на распределение `source` - источник трафика по количеству кликов  - построим круговую диаграмму;
        - посчитаем сколько денег потратили? Всего и на каждый источник;
        - сколько стоило привлечение одного пользователя из каждого источника?
        - Сформируем выводы.  
    
    - Датасет **user_source.csv**:
        - Есть ли в наборе данных отсутствующие признаки, изучим средние значения и стандартные отклонения;
        - Посмотрим на распределение количества пользователей по `source` (источник, с которого пришёл пользователь, установивший приложение) - построим круговую диаграмму;    
        - Сформируем выводы.

    - **по всем датасетам**:
        - посмотрим какие события совершают (`event` из events.csv) в зависимости от источника привлечения (`source` из user_source.csv), возможно появление интересного вывода (но это не точно) - построим круговую диаграмму.


4. Анализ данных по затратам
    - сколько денег потратили? Всего / на каждый источник / по времени
    - сколько стоило привлечение одного покупателя из каждого источника?
  
  
5. Анализ данных по предполагаемым продажам, расчет показателей
    - когда люди будут покупать (смотреть рекламу)?
    - сколько раз будут покупать за период (сессию)?
    - какой средний чек от пользователя (средний чек просмотра)?
    - сколько денег будут приносить? (LTV)


4. Проверка статистических гипотез;
    - Проверить гипотезу различия времени прохождения уровня между пользователями, которые заканчивают уровень через реализацию проекта, и пользователями, которые заканчивают уровень победой над другим игроком;
    - формирование и проверка гипотез после исследовательского анализа.
    
5. Презентация;


6. Выводы и формирование рекомендаций.
    - на сколько окупятся расходы? (ROI)
    - рекомендации по формированию стратегии.
